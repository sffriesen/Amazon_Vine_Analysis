```python
# Add dependencies
import pandas as pd
```

```python
# Read in csv file
vine_df = pd.read_csv("vine_table/vine_table.csv")
vine_df.head()
```

```python
# Create a new df to retrieve all the rows where the total_votes count is greater than or equal to 20
# new_df = 
new_df = vine_df[vine_df["total_votes"] >= 20]
new_df
```

```python
# Filter the new df and create a new one to retrieve all rows where 
# the number of helpful_votes divided by total_votes is equal to or greater than 50%

ratio_column = new_df["helpful_votes"] / new_df["total_votes"]
new_df["vote_ratio"] = ratio_column

filtered_df = new_df[new_df["vote_ratio"] > .5]
filtered_df
```

```python
# Filter the df and create a new one that retrieves all rows where 
# a review was written as part of the Vine program (unpaid), vine == "Y"
unpaid_df = filtered_df[filtered_df["vine"] == "Y"]
unpaid_df.head()
```

```python
# Filter the df and create a new one that retrieves all rows where 
# a review was not written as part of the Vine program (unpaid), vine == "N"
paid_df = filtered_df[filtered_df["vine"] == "N"]
paid_df.head()
```

```python
# Determine the total number of reviews
total_reviews = filtered_df["review_id"].count()
print(f"Total Reviews: {total_reviews}")
```

```python
# Determine the number of 5-star reviews
five_stars = filtered_df[filtered_df["star_rating"] == 5.0]

five_star_reviews = five_stars["star_rating"].count()
print(f"Five Star Reviews: {five_star_reviews}")
```

```python
# Determine the percentage of 5-star reviews for the two types of review (paid vs unpaid)

unpaid_df_count = unpaid_df["star_rating"].count()
unpaid_df_count

unpaid_five_stars = unpaid_df[unpaid_df["star_rating"] == 5.0]
unpaid_five_stars_count = unpaid_five_stars["star_rating"].count()
unpaid_five_stars_count

paid_df_count = paid_df["star_rating"].count()
paid_df_count

paid_five_stars = paid_df[paid_df["star_rating"] == 5.0]
paid_five_stars_count = paid_five_stars["star_rating"].count()
paid_five_stars_count

# (Unpaid Five-Star Reviews / Total Five-Star Reviews)
a = (unpaid_five_stars_count / five_star_reviews)
percentage_a = "{:.02%}".format(a)

# (Paid Five-Star Reviews / Total Five-Star Reviews)
b = (paid_five_stars_count / five_star_reviews)
percentage_b = "{:.02%}".format(b)

# (Unpaid Five-Star Reviews / Total Unpaid Reviews)
c = (unpaid_five_stars_count / unpaid_df_count)
percentage_c = "{:.02%}".format(c)

# (Paid Five-Star Reviews / Paid Reviews)
d = (paid_five_stars_count / paid_df_count)
percentage_d = "{:.02%}".format(d)

# unpaid_five_stars_count = five_stars["star_rating"].count()
print(f"(Unpaid Five-Star Reviews / Total Five-Star Reviews): {percentage_a}")
print(f"(Paid Five-Star Reviews / Total Five-Star Reviews): {percentage_b}")
print(f"(Unpaid Five-Star Reviews / Total Unpaid Reviews): {percentage_c}")
print(f"(Paid Five-Star Reviews / Paid Reviews): {percentage_d}")
```
